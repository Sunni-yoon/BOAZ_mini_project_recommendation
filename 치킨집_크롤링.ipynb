```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys # enter 치는 명령어
import pandas as pd 
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

from webdriver_manager.chrome import ChromeDriverManager



from selenium import webdriver
import time
from bs4 import BeautifulSoup
import re
```

```python


browser = webdriver.Chrome('C:/Users/blues/Downloads/chromedriver_win32 (1)/chromedriver') 
browser.get('https://www.yogiyo.co.kr/mobile/#/') #이렇게 하면 해당 url을 화면에 띄운다.
#search = browser.find_element(By.XPATH,'//*[@id="search"]/div/form/input')
search = browser.find_element(By.XPATH,'//*[@id="search"]/div/form/input')
search.click()


#검색어가 요기요가 기존위치를 설정하는데 이를 지우기
for _ in range(30):
    search.send_keys(Keys.ARROW_RIGHT)


for _ in range(40):
    search.send_keys(Keys.BACK_SPACE)


# 연세대학교 신촌 주변을 검색위치로 잡기
search.send_keys('서울특별시 서대문구 창천동 31-92 J-Tower 나동')
time.sleep(3)
# 검색창 찾고 누르기
search.send_keys(Keys.ENTER) 

result = []
result.append(0)
time.sleep(3)

# 치킨 카테고리를 클릭하기
browser.find_element(By.XPATH,'/html/body/div[6]/div/div[1]/div[1]/div/div/div/ul/li[5]').click()
time.sleep(3)





#3개의 음식점에서 리뷰를 가져오기
#이떄 n개의 식당에서 가졎오려면 아래에 range()에서 range(1,n+1)로 하면 된다.
for i in range(1,3+1):
    temp = []
    
    #i번째 음식점 클릭하기 아래에 i가 몇번쨰 치킨집을 의미함
    time.sleep(3)
    browser.find_element(By.XPATH,f'/html/body/div[6]/div/div[4]/div/div[2]/div[{i}]/div/table/tbody/tr/td[2]/div/div[1]').click()
    time.sleep(3)

    
    

    
    #음식점 홈페이지에서 리뷰보기를 클릭하기
    browser.find_element(By.XPATH,'/html/body/div[6]/div[2]/div[1]/ul/li[2]/a').click()
    time.sleep(3)


 

    AAA = int(browser.find_element(By.XPATH,'/html/body/div[6]/div[2]/div[1]/ul/li[2]/a/span').text)

    if AAA >100:
        
        #음식점 이름을 저장하기
        name = browser.find_element(By.XPATH,'/html/body/div[6]/div[2]/div[1]/div[1]/div[1]/span').text
        temp.append(name)
        
        
        start = 1
        #리뷰 한 음식점에서 100개를 가져옴
        #얘를 들어 200개를 가져오고 싶다면 아래 while에서 while <= 200으로 하면됨
        while len(temp) <= 100: 
            AAA = browser.find_elements(By.CSS_SELECTOR,'.list-group-item.star-point.ng-scope')
            for i in range(0,9+1):
                temp.append({start + i : AAA[i].get_attribute('innerText')})
            time.sleep(3)
            start += 10 

            browser.find_element(By.CSS_SELECTOR,"body").send_keys(Keys.END) # end키를 통해 스크롤을 맨 아래로 내림 
            time.sleep(1)

        # 더보기 클릭
            more_click = browser.find_element(By.CSS_SELECTOR,'li.list-group-item.btn-more')
            more_click.click()
            time.sleep(3)
        result.append(temp)
    #다시 치킨집 리스트로 이동하기
    browser.back() 
    time.sleep(3)

browser.close()



```
