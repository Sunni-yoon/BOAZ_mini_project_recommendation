```python
import re

temp_2 = df['리뷰']

def extract_word(text):
    hangul = re.compile('[^가-힣]') 
    result = hangul.sub(' ', text) 
    return result

print(extract_word(temp_2[10]))
```

```python
df['리뷰'] = df['리뷰'].apply(lambda x:extract_word(x))
df['리뷰'].unique()
```

```python
from konlpy.tag import Okt

okt = Okt()
words = " ".join(df['리뷰'].tolist())
words = okt.morphs(words,stem=True)
```

```python
remove_one_word = [x for x in words if len(x)>1 or x=="닭" or x == "맛"]
len(remove_one_word)
```

```python
from collections import Counter
frequent = Counter(remove_one_word).most_common()
print(frequent, end = "\n")
```

```python
with open('/content/drive/MyDrive/BOAZ-Mini-Project1/stopwords.txt', 'r',  encoding='cp949') as f:
    list_file = f.readlines()
stopwords = list_file[0].split(",")
remove_stopwords = [x for x in remove_one_word if x not in stopwords]
len(remove_stopwords)
```

```python
frequent = Counter(remove_one_word).most_common()
print(frequent, end = "\n")
```

```python
minimum_count = 3
more_than_one_time= []
for i in range(len(remove_stopwords)):
    tmp = remove_stopwords[i]
    if remove_stopwords.count(tmp) >= minimum_count:
        more_than_one_time.append(tmp)
```
